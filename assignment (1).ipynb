```python
import numpy as np
transactions=np.array([
    [1200,1300,1250,1400,1350,1500],
    [1100,1150,1200,1250,1300,1350],
    [1000,1050,1100,1150,1200,1250],
    [1500,1550,1600,1650,1700,1750]
])
total_per_branch=transactions.sum(axis=1)
print(total_per_branch)
highest_total_transactions=np.argmax(total_per_branch)
print(highest_total_transactions)
average_monthly_transactions=transactions.mean(axis=0)
reshaped=transactions.reshape(3,8)
print(average_monthly_transactions)
#implication: reshaping changes the structure but not the data.It's useful for reformatting input for modells or visualisations but it loses the original branch meaning. 
```

```python
routes=['Nairobi_Kisumu','Mombasa_Nairobi','Eldoret_Kakamega','Nakuru_Kericho','Kisii_Migori','Nairobi _Eldoret','Kitale_Bungoma','Garissa_Wajir','Lam_Malindi','Thika_Muranga']
print("All routes:",routes)
routes.append('Isiolo.Marsabit')
routes.remove('Nairobi_Kisumu')
print ("Updated routes:",routes)
routes.sort()
routes.reverse()
count_N=sum(route.startswith('N')for route in routes)
print("Sorted and reversed routes routes:",routes)
print("Routes starting with 'N':",count_N)
```

```python
electronics={"Laptop":15,"Phone":7,"Tablet":34,"Headphones":5,"Chargers":60}
print("electronics:",electronics)
electronics["Woofer"]=10
electronics["Phone"]=16
print("Added and updated poducts:",electronics)
def low_stock_items(electronics):
    return{item:quantity for item,quantity in electronics.items() if quantity<10}
    low_stock=low_stock_items(electronics)
    print("Low stock items:",low_stock)
    for item,quantity in low_stock.items():
        print(f"{item}:{quantity}")
del electronics["Headphones"]
print("deleted headphones:",electronics)
for item,quantity in electronics.items():
    print(f"{item}:{quantity}")
```
